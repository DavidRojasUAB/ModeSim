```python
import numpy as np
import matplotlib.pyplot as plt
```

# Bifurcacions

Quan considerem un sistema dinàmic discret $x_{n+1}=f_{\lambda}(x)$ on la funció iteració $f_{\lambda}$ depèn d'un paràmetre $\lambda$ de manera contínua, és natural preguntar-se si en variar $\lambda$ els punts fixos i òrbites periòdiques (o altres objectes) canvien en quantitat o caràcter. Quan un sistema presenta un canvi d'aquest tipus se'n diu *bifurcació* i el valor del paràmetre en el qual es dóna és un *paràmetre de bifurcació*.

Sigui $f_{\lambda}(x)$ una funció contínua i diferenciable amb continuitat respecte $x$ i $\lambda$, a partir d'ara la pensem com a funció en diverses variables $f(x,\lambda):=f_{\lambda}(x)$. Suposem que per a un cert valor del paràmetre $\lambda=\lambda^*$ tenim un punt fix $x^*=x^*(\lambda^*)$. Ens preguntem quan aquest punt fix segueix existint en moure lleugerament el paràmetre $\lambda\approx\lambda^*$. En altres paraules, busquem si existeix $x^*(\lambda)$ punt fix de $f(x,\lambda)$ satisfent
$$f(x^*(\lambda),\lambda)-x^*(\lambda)=0.$$
Això ens ha definit un equació implícita. Si definim $g(x,\lambda)=f(x,\lambda)-x$, tenim l'equació implícita $g(x,\lambda)=0$. Si $\frac{\partial}{\partial x}g(x^*,\lambda^*)\neq 0$ aleshores el Teorema de la Funció Implícita ens assegura que existeix una funció $x^*(\lambda)$ contínua prop de $\lambda=\lambda^*$ de tal manera que $g(x^*(\lambda),\lambda)=0$ per a tot $\lambda\approx\lambda^*$. En altres paraules, existeix un punt fix de $f(x,\lambda)$ per $\lambda\approx\lambda^*$.

- Si $\frac{\partial}{\partial x} f(x^*,\lambda^*)=1$ pot passar que $\lambda^*$ sigui paràmetre de bifurcació (per exemple, el punt fix pot desaparèixer).
- Si $\frac{\partial}{\partial x} f(x^*,\lambda^*)\neq 1$ el punt fix $x^*(\lambda)$ existeix per $\lambda\approx\lambda^*$.

## Tipus de bifurcacions

A continuació mostrem les bifurcacions més usuals en sistemes dinàmics discrets de dimensió 1.

### Bifurcació de sella-node
Dos punts fixos d'estabilitat oposada col·lisionen i desapareixen.

**Condicions:** $\frac{\partial}{\partial x}f(x^*,\lambda^*)=1$, $\frac{\partial^2}{\partial x^2}f(x^*,\lambda^*)\neq 0$ i $\frac{\partial}{\partial \lambda}f(x^*,\lambda^*)\neq 0$.

**Exemple:** $f(x,\lambda)=\lambda+x+x^2$ a $(x^*,\lambda^*)=(0,0)$.

```python
# Funció f(x, lambda)
def f(x, lam):
    return lam + x + x**2

# Diagonal y = x
def diagonal(x):
    return x

lam1 = -0.2
lam2 = 0.0
lam3 = 0.2
x = np.linspace(-1,1,400)
lam = np.linspace(-1,0,400)

fig, axes = plt.subplots(1, 2, figsize=(12,5))

axes[0].plot(x, f(x, lam1), label='$\lambda = -0.2$')
axes[0].plot(x, f(x, lam2), label='$\lambda = 0.0$')
axes[0].plot(x, f(x, lam3), label='$\lambda = 0.2$')
axes[0].plot(x, diagonal(x), 'k--', label='$y=x$')
axes[0].axhline(0, color='grey', linewidth=1)  # eix x
axes[0].axvline(0, color='grey', linewidth=1)  # eix y
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Bifurcació Sella-Node a $\lambda=0$')
axes[0].legend()
axes[0].axis(True)

axes[1].plot(lam,-np.sqrt(-lam), 'k-', label='Punts fixos atractors',linewidth=2)
axes[1].plot(lam,np.sqrt(-lam), 'k--', label='Punts fixos repulsors',linewidth=2)
axes[1].axhline(0, color='grey', linewidth=1)  # eix x
axes[1].axvline(0, color='grey', linewidth=1)  # eix y
axes[1].set_xlabel('$\lambda$')
axes[1].set_ylabel('x')
axes[1].set_title('Diagrama de bifurcació Sella-Node')
axes[1].set_xlim(-1, 1)  # eixos centrats
axes[1].set_ylim(-2, 2)
axes[1].legend()
axes[1].axis(True)

plt.show()
```

### Bifurcació de Pitchfork
Un punt fix canvia d'estabilitat i apareixen dos punts fixos que hereten l'estabilitat del punt fix original.

**Condicions:** $\frac{\partial}{\partial x}f(x^*,\lambda^*)=1$, $\frac{\partial^2}{\partial x^2}f(x^*,\lambda^*)=0$, $\frac{\partial^3}{\partial x^3}f(x^*,\lambda^*)\neq 0$ i $\frac{\partial^2}{\partial \lambda\partial x}f(x^*,\lambda^*)\neq 0$.

**Exemple:** $f(x,\lambda)=\lambda x+x-x^3$ a $(x^*,\lambda^*)=(0,0)$.

```python
# Funció f(x, lambda)
def f(x, lam):
    return lam*x + x - x**3

# Diagonal y = x
def diagonal(x):
    return x

lam1 = -0.2
lam2 = 0.0
lam3 = 0.2
x = np.linspace(-1,1,400)
lamda_span_1 = np.linspace(-1,0,200)
lamda_span_2 = np.linspace(0,1,200)

fig, axes = plt.subplots(1, 2, figsize=(12,5))

axes[0].plot(x, f(x, lam1), label='$\lambda = -0.2$')
axes[0].plot(x, f(x, lam2), label='$\lambda = 0.0$')
axes[0].plot(x, f(x, lam3), label='$\lambda = 0.2$')
axes[0].plot(x, diagonal(x), 'k--', label='$y=x$')
axes[0].axhline(0, color='grey', linewidth=1)  # eix x
axes[0].axvline(0, color='grey', linewidth=1)  # eix y
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Bifurcació Pitchfork a $\lambda=0$')
axes[0].legend()
axes[0].axis(True)

axes[1].plot(lamda_span_2,-np.sqrt(lamda_span_2), 'k-', label='Punts fixos atractors',linewidth=2)
axes[1].plot(lamda_span_2,np.sqrt(lamda_span_2), 'k-',linewidth=2)
axes[1].plot(lamda_span_2,0*lamda_span_2,'k--', label='Punts fixos repulsors',linewidth=2)
axes[1].plot(lamda_span_1,0*lamda_span_2,'k-',linewidth=2)
axes[1].axhline(0, color='grey', linewidth=1)  # eix x
axes[1].axvline(0, color='grey', linewidth=1)  # eix y
axes[1].set_xlabel('$\lambda$')
axes[1].set_ylabel('x')
axes[1].set_title('Diagrama de bifurcació Pitchfork')
axes[1].set_xlim(-1, 1)  # eixos centrats
axes[1].set_ylim(-2, 2)
axes[1].legend()
axes[1].axis(True)

plt.show()
```

### Bifurcació transcrítica
Dos punts fixos d'estabilitat oposada es creuen i intercanvien l'estabilitat.

**Condicions:** $\frac{\partial}{\partial x}f(x^*,\lambda^*)=1$, $\frac{\partial^2}{\partial x^2}f(x^*,\lambda^*)\neq 0$, $\frac{\partial}{\partial \lambda}f(x^*,\lambda^*)=0$ i $\frac{\partial^2}{\partial \lambda\partial x}f(x^*,\lambda^*)\neq 0$.

**Exemple:** $f(x,\lambda)=\lambda x(1-x)$ a $(x^*,\lambda^*)=(0,1)$.

```python
# Funció f(x, lambda)
def f(x, lam):
    return lam*x*(1-x)

# Diagonal y = x
def diagonal(x):
    return x

lam1 = 0.8
lam2 = 1.0
lam3 = 1.4
x = np.linspace(-0.5,1,400)
lambda_span_1 = np.linspace(0.01,1,200)
lambda_span_2 = np.linspace(1,2,200)

fig, axes = plt.subplots(1, 2, figsize=(12,5))

axes[0].plot(x, f(x, lam1), label='$\lambda = 0.8$')
axes[0].plot(x, f(x, lam2), label='$\lambda = 1.0$')
axes[0].plot(x, f(x, lam3), label='$\lambda = 1.4$')
axes[0].plot(x, diagonal(x), 'k--', label='$y=x$')
axes[0].axhline(0, color='grey', linewidth=1)  # eix x
axes[0].axvline(0, color='grey', linewidth=1)  # eix y
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Bifurcació transcrítica a $\lambda=1$')
axes[0].legend()
axes[0].axis(True)

axes[1].plot(lambda_span_2,(lambda_span_2-1)/lambda_span_2, 'k-', label='Punts fixos atractors',linewidth=2)
axes[1].plot(lambda_span_1,(lambda_span_1-1)/lambda_span_1, 'k--', label='Punts fixos repulsors',linewidth=2)
axes[1].plot(lambda_span_2,0*lambda_span_2,'k--',linewidth=2)
axes[1].plot(lambda_span_1,0*lambda_span_2,'k-',linewidth=2)

axes[1].axhline(0, color='grey', linewidth=1)  # eix x
axes[1].axvline(0, color='grey', linewidth=1)  # eix y
axes[1].set_xlabel('$\lambda$')
axes[1].set_ylabel('x')
axes[1].set_title('Diagrama de bifurcació transcrítica')
axes[1].set_xlim(0, 2)  # eixos centrats
axes[1].set_ylim(-2, 2)
axes[1].legend()
axes[1].axis(True)

plt.show()
```

### Bifurcació de duplicació de període
Un punt fix estable passa a ser inestable i apareix una òrbita 2-periòdica estable.

**Condicions:** $\frac{\partial}{\partial x}f(x^*,\lambda^*)=-1$ i $\frac{\partial^2}{\partial \lambda\partial x}f(x^*,\lambda^*)\neq 0$.

**Exemple:** $f(x,\lambda)=\lambda x(1-x)$ a $(x^*,\lambda^*)=(2/3,3)$.

```python
# Funció f(x, lambda)
def f(x, lam):
    return lam*x*(1-x)

def f2(x, lam):
    return f(f(x,lam),lam)

def op1(lam):
    return (1+lam+np.sqrt(lam**2-2*lam-3))/(2*lam)

def op2(lam):
    return (1+lam-np.sqrt(lam**2-2*lam-3))/(2*lam)
    
# Diagonal y = x
def diagonal(x):
    return x

lam1 = 2.8
lam2 = 3.0
lam3 = 3.2
x = np.linspace(0,1,400)
lambda_span_1 = np.linspace(1,3,200)
lambda_span_2 = np.linspace(3,3.4,200)

fig, axes = plt.subplots(1, 3, figsize=(18,5))

axes[0].plot(x, f(x, lam1), label='$\lambda = 2.8$')
axes[0].plot(x, f(x, lam2), label='$\lambda = 3.0$')
axes[0].plot(x, f(x, lam3), label='$\lambda = 3.2$')
axes[0].plot(x, diagonal(x), 'k--', label='$y=x$')
axes[0].axhline(0, color='grey', linewidth=1)  # eix x
axes[0].axvline(0, color='grey', linewidth=1)  # eix y
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Bifurcació duplicació de període a $\lambda=3$')
axes[0].legend()
axes[0].axis(True)

axes[1].plot(x, f2(x, lam1), label='$\lambda = 2.6$')
axes[1].plot(x, f2(x, lam2), label='$\lambda = 3.0$')
axes[1].plot(x, f2(x, lam3), label='$\lambda = 3.4$')
axes[1].plot(x, diagonal(x), 'k--', label='$y=x$')
axes[1].axhline(0, color='grey', linewidth=1)  # eix x
axes[1].axvline(0, color='grey', linewidth=1)  # eix y
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('Bifurcació duplicació de període a $\lambda=3$')
axes[1].legend()
axes[1].axis(True)

axes[2].plot(lambda_span_1,(lambda_span_1-1)/lambda_span_1, 'k-', label='Punts fixos atractors',linewidth=2)
axes[2].plot(lambda_span_2,(lambda_span_2-1)/lambda_span_2, 'k--', label='Punts fixos repulsors',linewidth=2)
axes[2].plot(lambda_span_2,op1(lambda_span_2),'b-',label='Òrbita 2-periòdica atractora',linewidth=2)
axes[2].plot(lambda_span_2,op2(lambda_span_2),'b-',linewidth=2)
axes[2].axhline(0, color='grey', linewidth=1)  # eix x
axes[2].axvline(0, color='grey', linewidth=1)  # eix y
axes[2].set_xlabel('$\lambda$')
axes[2].set_ylabel('x')
axes[2].set_title('Diagrama de bifurcació transcrítica')
axes[2].set_xlim(2,4)  # eixos centrats
axes[2].set_ylim(-0.5, 1.5)
axes[2].legend()
axes[2].axis(True)

plt.show()
```
