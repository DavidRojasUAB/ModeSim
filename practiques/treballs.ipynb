# Llista de treballs de Modelització i Simulació

## Treball 1. Modelització d'un partit de tennis.

**Motivació i context.** Un partit de tennis té una estructura jeràrquica molt clara: punts -> jocs -> sets -> partit. La pregunta clau és: si el jugador A guanya cada punt amb probabilitat constant $p$, quina és la probabilitat que guanyi el partit sencer? La resposta no és simplement $p$, i la diferència revela un mecanisme amplificador molt potent. A més, la història mostra una clara diferència entre la probabilitat de guanyar un punt de servei envers a guanyar un punt restant, la qual cosa modifica encara més la probabilitat de guanyar el partit.

**Descripció del problema.** Elaboreu un model per a simular l'evolució d'un partit de tennis fent servir cadenes de Markov amb l'objectiu de predir la probabilitat de victòria de cadascun dels jugadors i la durada mitjana del partit.

**Simulació del partit.** Preneu dades d'enfrontaments entre dos jugadors i calculeu la probabilitat de victòria del punt per servei i set. Feu servir aquestes probabilitats per a simular la cadena de Markov.

# Treball 2. Propagació d'epidèmies: el model SIR discret

**Motivació i context.** Des de la pandèmia de la COVID-19, els models epidemiològics han passat a ser eines de política pública. El model SIR (Susceptible-Infectat-Recuperat) és la peça fonamental d'epidemiologia matemàtica: simple, elegant i molt precís per a brots en poblacions homogènies. En la seva versió discreta, s'integra perfectament en el marc de la dinàmica discreta bidimensional.

**Descripció del problema.** Formuleu el model discret SIR derivant i interpretant el nombre reproductiu bàsic $R_0$: el nombre mitjà de nous infectats que genera un individu infectat en una població totalment susceptible. Describiu l'efecte $R_0<1$ envers a $R_0>1$ en l'epidèmia. Trobeu tots els punts fixos del sistema i analitzeu la seva estabilitat.

**Simulació numèrica.** Implementeu el model i simuleu per a diversos valors de $R_0$. Busqueu dades d'un brot real i estimeu els paràmetres per mínims quadrats. Compareu la predicció del model amb les dades.

# Treball 3. PageRank: com ordena Google les pàgines web

**Motivació i context.** L'algorisme PageRank, desenvolupat per Larry Page i Sergey Brin el 1998, va ser la innovació que va posar Google al capdavant dels motors de cerca. La idea és brillantment simple: una pàgina web és important si altres pàgines importants apunten cap a ella. Formalment, el rang de cada pàgina és la probabilitat que un "surfista aleatori" es trobi en aquella pàgina en el límit del temps.

**Descripció del problema.** Creeu una xarxa petita (50 nodes), representant-la com un graf dirigit i visualitzeu-la. Calculeu la matriu de transició a partir de la probabilitat per a cada pàgina de visitar cadascuna de les seves pàgines veïnes. Estableix un protocol per a les pàgines sense enllaços sortints i afegiu un factor $d\in(0,1)$ d'esmorteïment que modeli la rigurositat de seguir un enllaç envers a saltar a una pàgina aleatòria. La distribució PageRank $\pi$ satisfarà ser la distribució estacionària de la cadena de Markov. Analitzeu la convergència en funció del paràmetre d'esmorteïment $d$.

**Simulació.** Analitzeu quines pàgines obtenen el PageRank més alt i experimenteu afegint o eliminant alguns enllaços clau. Podeu fer servir també una subxarxa extreta de la Wikipedia (usant l'API de MediaWiki) en comptes d'una creada per vosaltres.


# Treball 4. Dinàmica de quotes en un mercat oligopolista

**Motivació i context.** En mercats com les telecomunicacions, la banca o les aerolínies, un nombre reduït d'empreses es reparteix la clientela. Cada mes, alguns clients canvien de proveïdor. Si es coneixen les probabilitats de canvi (churn rates), es pot modelar l'evolució de les quotes de mercat com una cadena de Markov i predir l'equilibri a llarg termini, el temps per arribar-hi, i l'efecte de campanyes de fidelització.

**Descripció del problema.** Busqueu dades de quotes de mercat per a un sector concret a Espanya: per exemple, les quotes trimestrals de les grans operadores de telefonia mòbil (Movistar, Orange, Vodafone, MásMóvil). Estimeu la matriu de transició a partir de les variacions de quotes observades. Verifiqueu les hipòtesis d'existència d'una única distribució estacionària i calculeu-la. 

**Simulació.** Analitzeu l'efecte de que una empresa llanci una campanya de retenció que redueixi la seva taxa de fugida. Estudieu l'anomenat Valor d'un client (Customer Lifetime Value).

# Treball 5. Estudi de la dinàmica caòtica de l'aplicació logística.

**Motivació i context.** L'equació logística $x_{n+1}=\lambda x_n (1-x_n)$ és una de les equacions més estudiades de les matemàtiques del s.XX. En aparença trivial, amaga una riquesa extraordinària: amb un sol paràmetre $\lambda$, exhibeix convergència a punt fix, oscil·lació periòdica, doblaments de període en cascada, i caos ple. La seva anàlisi va revelar universalitat: constants numèriques que apareixen en sistemes físics completament diferents.

**Descripció del problema.** Quan el paràmetre $\lambda>4$ la dinàmica a l'interval $[0,1]$ deixa de ser invariant. Estudieu com el conjunt invariant per la dinàmica queda reduït a un conjunt de Cantor i aquesta és conjugada a la dinàmica simbòlica de l'aplicació *shift* en l'espai de seqüències infinites $\{0,1\}^{\mathbb{N}}$. Demostreu que aquesta aplicació és caòtica en el sentit de Devaney.

**Simulació.** Fent servir també dinàmica simbòlica es pot codificar l'òrbita d'un punt a $[0,1]$ quan $\lambda<4$, diferenciant entre *estar a l'esquerra* del màxim $(0)$, a la *dreta del màxim* $(1)$ o ser *el màxim* $(C)$. Fent simulacions trobeu una aproximació per a l'entropia topològica de l'aplicació en funció de $\lambda$.

**Referència.** Morris W. Hirsch, Stephen Smale, Robert L. Devaney. Differential Equations, Dynamical Systems & An Introduction to Chaos.



# Treball 6. El model de Schelling: segregació urbana espontània

**Motivació i context.** Thomas Schelling, premi Nobel d'Economia el 2005, va proposar el 1971 un model sorprenent: agents amb una tolerància moderada cap a veïns d'un altre grup produeixen, per la seva interacció local, una segregació espacial molt superior a la que la tolerància individual semblaria justificar.

El model és un cas paradigmàtic de comportament emergent: el resultat col·lectiu és qualitativament diferent de les intencions individuals, i il·lustra per què la intuïció falla en sistemes complexos.

**Descripció del problema.** Distribuïu aleatòriament tres tipus de cel·les en una graella $N\times N$: agents vermells, agents blaus i cel·les buides. Un agent es considera satisfet si la proporció de veïns del seu tipus entre els seus veïns ocupats és $\geq \tau$ (llindar de tolerància). En cas contrari, l'agent està insatisfet i vol moure's.

**Simulació.** Estableix el protocol de la dinàmica de moviment i repetiu-lo fins que tots els agents estiguin satisfets (equilibri) o fins a un cert màxim d'iteracions. Mesureu la segregació: la proporció mitjana de veïns del mateix tipus per sobre del 50% esperat (si fos aleatòri). Feu un estudi de com afecta el llindar de tolerància. Interpreteu els resultats: fins a quin punt el model captura fenòmens reals de segregació urbana? Quines simplificacions fa? Com es podria ampliar per incorporar preus del sòl, xarxes socials o discriminació activa? Estudieu què passa quan el llindar de tolerància depen del tipus d'agent (vermell o blau) i l'efecte d'introduir un tercer agent diferent als dos anteriors.
