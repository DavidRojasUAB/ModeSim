# Sisdemes dinàmics discrets lineals en diverses variables

Considerem l'equació en diferències

$$
\begin{cases}
    x_{n+1} & = 2 x_n-3 y_n\\
    y_{n+1} & = x_n - 2y_n.
\end{cases}
$$

Es tracta d'un sistema de primer ordre amb dues equacions lineals. Anem a resoldre el sistema de dues maneres: fent servir els mètodes lineals d'ordre superior d'una variable, i fent servir la teoria matricial.

### Reducció a una equació de segon ordre

De la primera equació tenim
$$ y_n = -\frac{1}{3}x_{n+1} + \frac{2}{3} x_n. $$
Canviant els índexos de la primera equació ($n\rightarrow n+1$) i fent servir l'expressió de $y_{n+1}$ de la segona equació:
$$ x_{n+2} = 2x_{n+1}-3y_{n+1} = 2x_{n+1} - 3x_n + 6 y_n.$$
Finalment substituint $y_n$ en aquesta última:
$$ x_{n+2} = 2x_{n+1}-3x_n+6(-\frac{1}{3}x_{n+1} + \frac{2}{3} x_n) = 2x_{n+1} - 3x_n - 2x_{n+1}+4x_n.$$
Així obtenim l'equació de segon ordre:
$$ x_{n+2} - x_n = 0.$$
El polinomi característic associat és $p(\lambda)=\lambda^2-1$ que té per arrels $\lambda_1=1$ i $\lambda_2= -1$ i, per tant, la solució general real és
$$ x_n = c_1+c_2(-1)^n, \quad c_1,c_2\in\mathbb{R}.$$
Per trobar $y_n$ tornem a fer servir la primera equació:
$$ y_n = -\frac{1}{3}x_{n+1} + \frac{2}{3} x_n = -\frac{1}{3}(c_1+c_2(-1)^{n+1})+\frac{2}{3}(c_1+c_2(-1)^n) = \frac{1}{3}c_1 + c_2(-1)^n.$$

### Mètode matricial


Escrivim $X_n = \begin{pmatrix} x_n \\ y_n \end{pmatrix}\in \mathbb{R}^2$ i $A = \begin{pmatrix} 2 & -3 \\ 1 & -2 \end{pmatrix}$. Llavors, el sistema original s'escriu com $$ X_{n+1} = A X_{n}$$ i podem interpretar el sistema com una generalització de les progressions geomètriques en el context vectorial. Ara la raó és una matriu, el terme general un vector i el producte és no commutatiu. Per inducció, la fórmula del terme general és
$$X_n = A^n X_0.$$
Doncs, resoldre el sistema passa per conèixer el càlcul de les potències d'una matriu.

## Teorema de Cayley-Hamilton

Donada una matriu quadrada $A\in\mathbb{C}^{d\times d}$, es defineix el seu polinomi característic com $p(\lambda) = \det(A-\lambda I)$. Es compleix que $$ p(A) = 0.$$ És a dir, tota matriu quadrada anul·la el seu polinomi característic.

Donat un polinomi $q(\lambda) = a_k\lambda^k+a_{k-1}\lambda^{k-1}+\cdots + a_1\lambda + a_0$, l'avaluació $q(A)$ es defineix com $$q(A) = a_k A^k+a_{k-1} A^{k-1}+\cdots + a_1 A + a_0 I$$ on $I$ és la matriu identitat de dimensió $d\times d$.

Així, el Teorema de Cayley-Hamilton estableix una relació entre $A^d$ i les potències anteriors de $A$. En efecte, si $p(\lambda) = \lambda^d+a_{d-1}\lambda^{d-1}+\cdots+a_1\lambda+a_0$ aleshores $$ A^d = -a_{d-1} A^{d-1}-\dots - a_1A - a_0 I.$$ Si multipliquem per $A^n$ pes l'esquerra: $$ A^{n+d} = -a_{d-1} A^{n+d-1}-\dots - a_1A^{n+1} - a_0 A^n.$$ 


Moltes vegades és possible calcular les potències de $A$ a partir d'aquesta llei de recurrència. Tornem a l'exemple inicial:

Considerem $X_{n+1} = A X_n, \quad  A = \begin{pmatrix} 2 & -3 \\ 1 & -2 \end{pmatrix}$. Tenim $p(\lambda) = \lambda^2-1$. Pel Teorema de Cayley-Hamilton, $A^2=I$ i, per tant, $$ A^n = \begin{cases} A & \text{ si } n \text{ senar }\\ I & \text{ si } n \text{ parell }\end{cases}$$

Així, la solució del sistema amb condició inicial $X_0 = \begin{pmatrix} x_0 \\ y_0 \end{pmatrix}$ és $$X_n = A^nX_0 = \begin{cases} AX_0 & \text{ si } n \text{ senar }\\ X_0 & \text{ si } n \text{ parell }\end{cases} = \begin{cases} \begin{pmatrix} 2x_0-3y_0 \\ x_0-2y_0 \end{pmatrix} & \text{ si } n \text{ senar }\\ \begin{pmatrix} x_0 \\ y_0 \end{pmatrix} & \text{ si } n \text{ parell }\end{cases}$$
